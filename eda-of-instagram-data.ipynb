## Extracting data From "business_address_json" 

```python
import json
```

```python
data['business_address_json'] = data['business_address_json'].apply(lambda x: '{}' if pd.isna(x) or isinstance(x, float) else x)
```

```python
data['location_details'] = data['business_address_json'].apply(json.loads)
```

```python
df = pd.concat([data, pd.json_normalize(data['location_details'])], axis=1)
```

```python
df.drop(columns=['location_details'], inplace=True)

df.head()
```
